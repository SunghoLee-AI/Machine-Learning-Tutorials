```python
# Load necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
filename = 'https://library.startlearninglabs.uw.edu/DATASCI410/Datasets/JitteredHeadCount.csv'
headcount_df = pd.read_csv(filename)
print(headcount_df.shape)
print(headcount_df.dtypes)
print(headcount_df.describe())
print(headcount_df.head(10))
print(headcount_df.tail(10))
```

```python
%matplotlib inline
fig = plt.figure(figsize=(6, 6)) # define plot area
ax = fig.gca() # define axis 
headcount_df.loc[:,['HeadCount', 'GameCode']].boxplot(by = 'GameCode', ax = ax)
ax.set_title('Box plot of HeadCount') # Give the plot a main title
ax.set_ylabel('HeadCount')# Set text for y axis
#ax.set_ylim(0.0, 50.0) # Set the limits of the y axis
```

```python
Correlation1=headcount_df[['HeadCount', 'TablesOpen', 'TablesOcc', 'TablesClosed']].corr()
Correlation1
```

```python
sns.heatmap(Correlation1, vmax=1.0, center=0,
            square=True, linewidths=.25, cbar_kws={"shrink": .5})
plt.title('Correlation matrix for numeric features')
plt.yticks(rotation='horizontal')
plt.xticks(rotation='vertical')
```
